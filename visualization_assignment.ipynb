```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plot - Cambios de temperatura

Este gráfico muestra como cambió la temperatura desde el 4/5/2025 hasta el 10/5/2025. Un gráfico de líneas es ideal para datos temporales porque permite visualizar tendencias y patrones de manera clara.

Elegí esta visualización porque facilita observar aumentos o disminuciones en un intervalo continuo.

```python
days = np.arange(4, 11)
temperature = [23, 19, 22, 20, 19, 20, 20]

plt.figure()
plt.plot(days, temperature, marker='o', linestyle='--', color='blue', label='Temperature')

plt.title('Temperatura 4/5/2025 - 10/5/2025')
plt.xlabel('Días')
plt.ylabel('Temperatura (°C)')

plt.savefig('imgs/temperatura_semana.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Bar Chart - Comparación de Ventas

Este gráfico de barras compara las ventas entre diferentes productos. Los gráficos de barras son útiles para comparar diferentes categorías.

Elegí esta visualización porque permite ver claramente las diferencias de magnitud entre cada categoría.

```python
products = ['A', 'B', 'C', 'D']
sales = [100, 150, 80, 200]

plt.figure()
plt.bar(products, sales, color='green')

plt.title('Ventas de productos')
plt.xlabel('Productos')
plt.ylabel('Ventas (x unidad)')

plt.savefig('imgs/ventas_productos.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Histogram - Distribución de Edades

Este histograma muestra la distribución de edades en un conjunto de datos. Los histogramas son útiles para entender cómo se distribuyen los datos.

Elegí esta visualización para observar la frecuencia y detectar patrones como la concentración o dispersión de los valores.

```python
ages = np.random.normal(30, 10, 100)

plt.figure()
plt.hist(ages, bins=10, color='red', edgecolor='black')

plt.title('Pirámide de población "X"')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')

plt.savefig('imgs/distribucion_edades.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Scatter Plot - Altura vs Peso

Este gráfico de dispersión muestra la relación entre la altura y el peso. Este tipo de gráfico es ideal para identificar correlaciones entre variables.

Elegí esta visualización para analizar si existe una relación entre ambas variables.

```python
height = np.random.randint(150, 190, 50)
weight = np.random.randint(50, 100, 50)

plt.figure()
plt.scatter(height, weight, color='yellow', marker='o')

plt.title('Altura vs Peso')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')

plt.savefig('imgs/altura_peso.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Box Plot - Notas de Exámenes

Este diagrama de caja resume las notas de exámenes utilizando cuartiles. Permite visualizar la mediana, la dispersión y posibles valores atípicos.

Elegí esta visualización porque ofrece un resumen estadístico claro y compacto de los datos.

```python
scores = [np.random.randint(50, 100, 30) for _ in range(3)]

plt.figure()
plt.boxplot(scores, tick_labels=['Matemáticas', 'Ciencias', 'Inglés'])

plt.title('Notas x materia')
plt.xlabel('Materias')
plt.ylabel('Notas')

plt.savefig('imgs/notas_examenes.png', dpi=300, bbox_inches='tight')
plt.show()
```
